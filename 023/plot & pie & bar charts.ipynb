```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.arange(-3, 3, 0.001)
```

نمایش اطلاعات جدول

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))

plt.legend(['A', 'B'], loc=1)

plt.show()
```

(Pie Chart) رسم نمودار دایره ای

```python
values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']
labels = ['Iran', 'USA', 'Russia', 'Euroup', 'China']
explods = [0.2, 0, 0.1, 0, 0]

plt.pie(values, colors=colors, labels=labels, explode=explods)
plt.show()
```

(Bar Chart) رسم نمودار میله ای

```python
values = [12, 35, 56, 5]
colors = ['r', 'g', 'b', 'm']
plt.bar(range(0, 4), values, color=colors)
plt.show()
```
