```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

CSV لود کردن فایل

```python
data = pd.read_csv('../Data/cars.csv')
data.head()
```

جدید Data Frame جدا کردن 4 ستون دلخواه و ساخت یک

```python
df = data[['Cylinders', 'CityMPG', 'HwyMPG', 'CombMPG']]
df.head()
```

Pair Plot رسم نمودار 

```python
sb.pairplot(df, hue='Cylinders', height=2.5)
```

رسم نمودار نقطه ای از مقایسه دو فیلد

```python
sb.scatterplot(x='Eng Displ', y='CombMPG', data=data)
```

y و x رسم نمودار نقطه ای مقایسه دو فیلد به همراه رسم نمودار میله ای از فیلد های

```python
sb.jointplot(x='Eng Displ', y='CombMPG', data=data)
```
