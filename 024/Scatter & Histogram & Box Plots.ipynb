```python
import numpy as np
import matplotlib.pyplot as plt
```

(Scatter Plot) نمودار نقطه ای

```python
x = np.random.randn(50)
y = np.random.randn(50)

plt.scatter(x, y)
plt.show()
```

(Histogram) نمودار هیستوگرام

```python
values = np.random.normal(10000, 1000, 10000)
plt.hist(values, 50)
plt.show()
```

(Box Plot) نمودار جعبه ای

![Getting Started](BoxPlot2.png)

.در این نمودار 50٪ داده ها بین چارک اول و سوم قرار دارند

.همچنین 25٪ داده ها کمتر از چارک اول و 75٪ داده ها کمتر از چارک سوم قرار دارند

.داده پرت محسوب می شوند Q1 - 1.5 * IQR داده های قبل از

.داده پرت محسوب می شوند Q3 + 1.5 * IQR داده های بعد از

```python
values = np.random.rand(100) * 100 - 40
high = np.random.rand(10) * 50 + 100
low = np.random.rand(10) * -50 - 100

data = np.concatenate((values, high, low))
plt.boxplot(data)
plt.show()
```
