````python
import pandas as pd
````

````python
r_cols = ['user_id', 'movie_id', 'rating']
rating = pd.read_csv('../Data/movies data/movies_data.txt', sep='\t', names=r_cols, usecols=range(3))
rating.head(10)
````

````python
m_cols = ['movie_id', 'title']
movies = pd.read_csv('../Data/movies data/movies_item.txt', sep='|', names=m_cols, usecols=range(2))
movies.head(10)
````

````python
ratings = pd.merge(movies, rating)
ratings.head(10)
````

````python
movieRating = ratings.pivot_table(index='user_id', columns='title', values='rating')
movieRating.head(10)
````

Find movies similar to a movie

Tips :

- The ```corrwith()``` function in Pandas computes pair-wise correlations between rows and columns of a dataframe with rows and columns of a series or dataframe.

- The ```dropna()``` method removes the rows that contains NULL values.

````python
starWarsRating = movieRating['Star Wars (1977)']

similarMovies = movieRating.corrwith(starWarsRating)
similarMovies = similarMovies.dropna()

similarMovies = similarMovies.sort_values(ascending=False)

df = pd.DataFrame(similarMovies)
df.head()
````
