Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.metrics import r2_score

np.random.seed(2)
```

Config Matplotlib

```python
def plt_config():
    sb.set()
    plt.xlabel('Page Speed')
    plt.ylabel('Purchase Amount')

    plt.legend(loc="upper right", bbox_to_anchor=(1, 1))

    plt.show()
```

Clear Coefficient of Determination

```python
def clear_coefficient_determination(value):
    value = round(value * 100, 3)
    print(f'{value}%')
```

Split Data

```python
data_count = 100

Percent_80 = int(data_count * 80 / 100)

print(f'Len 80% : {Percent_80}')
```

Create Random Data

```python
pageSpeed = np.random.normal(3.0, 1.0, data_count)
purchaseAmount = np.random.normal(50.0, 30.0, data_count) / pageSpeed
```

Split Train & Test Data

```python
trainX = pageSpeed[:Percent_80]
testX = pageSpeed[Percent_80:]

trainY = purchaseAmount[:Percent_80]
testY = purchaseAmount[Percent_80:]
```

Show Train & Test Data

```python
plt.scatter(trainX, trainY, c='b', label='Train Data')
plt.scatter(testX, testY, c='r', label='Test Data')
plt_config()
```

```python
train_x = np.array(trainX)
train_y = np.array(trainY)

test_x = np.array(testX)
test_y = np.array(testY)
```

Create Polynomial Fit

```python
p4 = np.poly1d(np.polyfit(train_x, train_y, 7))
print(p4)
```

Show Result

```python
xp = np.linspace(0, 7, 100)
axes = plt.axes()
axes.set_xlim([0, 7])
axes.set_ylim([0, 200])

plt.scatter(train_x, train_y, c='b', label='Train Data')
plt.scatter(test_x, test_y, c='r', label='Test Data')

plt.plot(xp, p4(xp), c='g', label='Predict')

plt_config()
```

Coefficient of Determination for Train Data

```python
r2_train = r2_score(trainY, p4(trainX))
clear_coefficient_determination(r2_train)
```

Coefficient of Determination for Test Data

```python
r2_test = r2_score(test_y, p4(test_x))
clear_coefficient_determination(r2_test)
```
