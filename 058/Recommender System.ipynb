```python
import pandas as pd
import numpy as np
```

```python
r_cols = ['user_id', 'movie_id', 'rating']
rating = pd.read_csv('../Data/movies data/movies_data.txt', sep='\t', names=r_cols, usecols=range(3))

m_cols = ['movie_id', 'title']
movies = pd.read_csv('../movies_item.txt', sep='|', names=m_cols, usecols=range(2))

ratings = pd.merge(movies, rating)

movieRating = ratings.pivot_table(index='user_id', columns='title', values='rating')
movieRating.head()
```

```python
starWarsRating = movieRating['Star Wars (1977)']

similarMovies = movieRating.corrwith(starWarsRating)
similarMovies = similarMovies.dropna()

# similarMovies = similarMovies.sort_values(ascending=False)

df = pd.DataFrame(similarMovies)
df.head()
```

```python
similarMovies.sort_values(ascending=False)
```

```python
movieStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movieStats.head()
```

```python
popularMovies = movieStats['rating']['size'] >= 100
movieStats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:15]
```

```python
df = movieStats[popularMovies].join(pd.DataFrame(similarMovies, columns=['similarity']))
# df.head()
```
