```python
import pandas as pd
import numpy as np
```

```python
r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('../Data/movies data/movies_data.txt', sep='\t', names=r_cols, usecols=range(3))
ratings.head(10)
```

```python
movieProperties = ratings.groupby('movie_id').agg({'rating':[np.size, np.mean]})
movieProperties.head(10)
```

```python
movieNumRatings = pd.DataFrame(movieProperties['rating']['size'])
movieNormalaizedNumRatings = movieNumRatings.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))
movieNormalaizedNumRatings.head(10)
```
