```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sb

sb.set()
```

```python
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchasesAmount = 100 - (pageSpeeds + np.random.normal(0.0, 0.1, 1000)) * 3
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchasesAmount)

print(f'slope : {slope}\nintercept : {intercept}\nr_value : {r_value}\np_value : {p_value}\nstd_err : {std_err}\n')
```

```python
def predict(x):
    return slope * x + intercept
```

```python
fitLine = predict(pageSpeeds)
plt.scatter(pageSpeeds, purchasesAmount)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
```
