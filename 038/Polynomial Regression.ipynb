```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import seaborn as sb

sb.set()
```

```python
np.random.seed(2)
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds
plt.scatter(pageSpeeds, purchaseAmount)
plt.show()
```

محاسبه رگرسیون مرتبه 4

یک تابع است p4() توجه کنید که

```python
x = np.array(pageSpeeds)
y = np.array(purchaseAmount)

p4 = np.poly1d(np.polyfit(x, y, 4))     # سومین ورودی این تابع مرتبه نمودار است
print(p4)
```

```python
xp = np.linspace(0, 7, 300)     # با استفاده از این دستور یک محدوده برای نمودار ایجاد می کنیم
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')     # است y که همان p4(x) و ورودی دوم x ورودی اول مقدار
plt.show()


```

پیش بینی مقدار 0.2

```python
print(p4(0.2))
```

محاسبه مقدار خطای نمودار

```python
r2 = r2_score(y, p4(x))
print(r2)
```
